```
import spacy
print(spacy.__version__)
# basic imports
import nltk

from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
nlp = spacy.load('en_core_web_sm')
```

```
words = ['run', 'runner', 'running', 'wolves','ran', 'easily', 'fairly']
```

```
pStem = PorterStemmer()
sStem = SnowballStemmer(language = 'english')
```

# Stemming is Very Basic (Porter worse than Snowball)



```
for word in words:
  print(word + " --------- " +pStem.stem(word))
```

```
for word in words:
  print(word + " --------- " +sStem.stem(word))
```

# Lemmatization (Generally Better)

```
doc1 = nlp('run runner running wolves ran easily fairly since they are the best')
```

```
for token in doc1:
  print(token.text, '\t', token.lemma_)
```
