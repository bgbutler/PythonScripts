```
# this may take some time

nlp = spacy.load(name = 'en_core_web_md')
```

```
doc1 = nlp(s1)
```

```
print(s1)
for token in doc1:
  print(token)
```

```
# do for S2
# web URLs are one token
doc2 = nlp(s2)
print(doc2)
for token in doc2:
  print(token)
```
