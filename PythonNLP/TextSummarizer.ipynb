```
doc = nlp(text)
```

```
tokens = [token.text for token in doc]
print(tokens)
```

```
punctuation
```

```
punctuation = punctuation + '\n'
punctuation
```

# Text Cleaning

```
word_freq = {}
stop_words = list(STOP_WORDS)

for word in doc:
  # remove punctuation and stop words
  if word.text.lower() not in stop_words:
    if word.text.lower() not in punctuation:
      # begin to build the dictionary
      if word.text not in word_freq.keys():
        word_freq[word.text] = 1
      else:
        word_freq[word.text] += 1
```

```
print(word_freq)
```

```
max_freq = max(word_freq.values())
```

```
max_freq
```

```
# normalize by dividing by the max freq
for word in word_freq.keys():
  word_freq[word] = word_freq[word]/max_freq
```

```
print(word_freq)
```

# Sentence Tokenization

```
sent_tokens = [sent for sent in doc.sents]
print(sent_tokens)
```

```
# score sentences based on values, make a dictionary
sent_score = {}
```

```
# fill the sentence score dictionary
for sent in sent_tokens:
  for word in sent:
    if word.text.lower() in word_freq.keys():
      if sent not in sent_score.keys():
        sent_score[sent] = word_freq[word.text.lower()]
      else:
        sent_score[sent] += word_freq[word.text.lower()]
```

```
print(sent_score)
```

# Find the Highest Scoring Sentences (Top 30%)

```
from heapq import nlargest
```

```
len(sent_score)
```

```
len(sent_score) * 0.3
```

```
# 32 lines is still too long, take top 10% (11 sentences)
summary = nlargest(n = 11, iterable = sent_score, key = sent_score.get)
print(summary)
```

```
# make it a single bucket
final_summary = [word.text for word in summary]
print(final_summary)
```

```
# get in clean paragraph format
summary_clean = " ".join(final_summary)
print(summary_clean)
```
