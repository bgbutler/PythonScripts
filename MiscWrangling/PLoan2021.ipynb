```python
# CAMPAIGN PERIOD
campaign_start = dt.datetime.strptime("2021-01-01","%Y-%m-%d")
campaign_end = dt.datetime.strptime("2021-03-01","%Y-%m-%d")

# LOAD MAIN FILE
PL_APPS = pd.read_csv('Eastern_ConsumerLoanApps.csv', low_memory=False)
PL_APPS.head()
```

```python
PL_APPS.shape
```

```python
# CALCULATE APPLICATION DATE
# IF SentToDecision IS BLANK, THEN USE AppEntryCompletedDate
# PL_APPS['AppEntryCompletedDate'] = pd.to_datetime(  PL_APPS['AppEntryCompletedDate'].astype(str) , format='%m/%d/%Y' )
# PL_APPS['SentToDecision'] = pd.to_datetime(PL_APPS['SentToDecision'].astype(str) , format='%m/%d/%Y %H:%M %p' )
PL_APPS['APPLICATION_DATE'] = [ PL_APPS['SentToDecision'][x] if str(PL_APPS['SentToDecision'][x])!='NaT' else PL_APPS['AppEntryCompletedDate'][x] for x in range(len(PL_APPS['SentToDecision'])) ]


PL_APPS['AppEntryCompletedDate'] = pd.to_datetime(PL_APPS['AppEntryCompletedDate'], format = '%m/%d/%Y')


```

```python
# CLEAN UP CIS
PL_APPS.CustomerNumber = [ re.sub("[^0-9]", "", str(x) ) for x in PL_APPS.CustomerNumber]
PL_APPS.CustomerNumber = [ int(x) if x != '' else np.NaN for x in PL_APPS.CustomerNumber]
```

```python
# CLEAN UP PRODUCT
PL_APPS['product'] = PL_APPS['product'].str.strip()
```

```python
# LIMIT TO RELEVANT ACCOUNTS
PL_APPSQ1 = PL_APPS.loc[    
    (PL_APPS.AppEntryCompletedDate>=campaign_start) &
    (PL_APPS.AppEntryCompletedDate<=campaign_end) &
    (PL_APPS['product'] == 'Personal Loan') &
    (PL_APPS['LastDecision'] != 'Cancelled') & 
    (PL_APPS['BookedYN'] == 'Yes')
]   

PL_APPSQ1.info()
```
