```python
# create a in footprint test
# create a condition to check

def footprint_col(colCheck):
    
    footprint = ['ME', 'VT', 'NH', 'MA', 'RI', 'CT', 'NY', 'NJ', 'PA']
    
    if colCheck in footprint:
        return('In')
    else:
        return('Out')
    
data['Footprint'] = data['State'].apply(footprint_col)
```
