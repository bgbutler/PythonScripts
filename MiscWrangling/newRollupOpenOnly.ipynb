```python
# standard libraries
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import os
import re

# plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# get the datetime library for date & time calcs
from datetime import datetime, timedelta
```

```python
# load the the main data sets
wideData = os.path.normpath('C:/Users/n846490/Documents/Python Scripts/SurvivalAnalysis/DataForRollupCleanState.csv')
rollupData = pd.read_csv(wideData)
```

```python
# create an open function
# this is applied to the closed column

def open(row):
    if row == 0:
        return int(1)
    else:
        return int(0)
```

```python
# now apply the function to create a new column called open

rollupData['Open'] = rollupData['Closed'].apply(open)

rollupData.head(10)
```

```python
# drop the unnamed column

rollupData.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
rollupData['OpenAccts'] = rollupData['Open'] * rollupData['NumProducts']
```

```python
rollupData.info()
```

```python
# df[df["location"] == "c"].squeeze()
# this is used to investigate a single customer
# has 7 open accounts and 7 closed accounts
# therefore through the customer life, has had 14 accounts

rollupData[rollupData['CustID'] == 'F100001'].squeeze()
```

```python
openAccts = rollupData.copy()
```

```python
# need a function to multiply the open column by each column

# first get a list of the column names
# there are 48 columns
# len(products)

keyProducts = ['Cdira', 'Checking', 'Credit_Card', 'Investments', 'Mortgage', 'Savings']

for prod in keyProducts:
    
    # create the column header
    newName = prod + 'Open'
    
    # create the values in the new column
    openAccts[newName] = openAccts['Open'] * openAccts[prod]
    
```

```python
openAccts.head(50)
```

```python
# perform the aggregation
groupedOpen = openAccts.groupby(['CustID', 'State', 'Channel', 'Age', 'Cleancuststart', 'Custyears' , 'Footprint'], as_index = False).sum()
```
