```python
# first create an empty dictionary
d = dict()

myFile = 'deck1UTF.txt'


# open the file
with open(myFile, 'r') as myFile:
    
    data = myFile.read()
    
    pattern = re.compile(r'\(3\..*?\)')
    
    result = pattern.split(data)
    
    totQuestions = len(result)
    
for r in range(len(result)):
    d[r] = result[r]
    
questionDf = pd.DataFrame(list(d.items()), columns=['Number', 'Question'])
    
print('Dictionary', d)
print('There are ', totQuestions, ' questions in the file.')
print('The length of the Dictionary is ', len(d))

```

```python
# drops the first line

questionDf.drop([0], axis = 0, inplace = True)

questionDf.head()
```

```python
# get the answers by splitting on the '(a)'

ans = r'\(A\)'

questionDf = questionDf['Question'].str.split(ans, expand=True)

questionDf.rename(columns = {0:'Question', 1:'Options', 2:'None'}, inplace = True)

# drop column 2 (renamed to None)
questionDf.drop('None', axis = 1, inplace = True)

questionDf.head()
```

```python
# get a look at the new df

# questionDf.to_csv('DeckQuestionsOptionsV1.csv')
```

```python
# split again on the 'If Choice'

choice = 'If choice'

questionDf[['Options', 'Choice']] = questionDf['Options'].str.split(choice, expand=True)

questionDf.head()
```

```python
# modify the text to be clean for each column
# set the choice to a single character

questionDf['Choice'] = questionDf['Choice'].str[1]

questionDf.head()
```

```python
questionDf.iloc[0,1]
```

```python
# check out the Options might need to add back the (A)

questionDf['Options'] = '(A) ' + questionDf['Options']

questionDf.head()
```

```python
questionDf.iloc[1,1]
```

```python
# first replace the \n with a space
# the new line character only appears in the options
    
questionDf['Options'] = questionDf['Options'].replace({r'\n': ''}, regex=True)

questionDf.head()
```

```python
# remove the answers and should be good
# the original regex is the following '[o ][o ][o ][o ]*'
# a better regex [\.,\? ][o,1 ][o,1 ][o,1 ][o,1 ]*  deals with issues like to operate

questionDf['Question'] = questionDf['Question'].replace({r'[\.,\? ][o,1 ][o,1 ][o,1 ][o,1 ]*': ' '}, regex=True)
```
